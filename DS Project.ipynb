```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
DataFrameCombined = pd.read_csv("men_women.csv")
```

```python
DataFrameCombined
```

```python
DataFrameCombined.fillna(DataFrameCombined.mean(), inplace=True)
DataFrameCombined
```

```python
dfmens = pd.DataFrame(columns=DataFrameCombined.columns)
dfmens =dfmens.append(DataFrameCombined[DataFrameCombined["SUBJECT"]=='MEN'],ignore_index=True)
dfmens
```

```python
dfwomen = pd.DataFrame(columns=DataFrameCombined.columns)
dfwomen =dfwomen.append(DataFrameCombined[DataFrameCombined["SUBJECT"]=='WOMEN'],ignore_index=True)
dfwomen
```

```python
dfmensfinal = pd.get_dummies(dfmens.LOCATION)
dfmensfinal.insert(44,"Time",dfmens["TIME"])
dfmensfinal.insert(45,"value",dfmens["Value"])
dfmensfinal
```

```python
dfwomensfinal = pd.get_dummies(dfwomen.LOCATION)
dfwomensfinal.insert(44,"Time",dfwomen["TIME"])
dfwomensfinal.insert(45,"value",dfwomen["Value"])
dfwomensfinal
```

```python
Ywomens = dfwomensfinal['value']
dfwomensfinal.pop('value')
Xwomens = dfwomensfinal
```

```python
Ymens = dfmensfinal['value']
dfmensfinal.pop('value')
Xmens = dfmensfinal
```

```python
MenReg = LinearRegression()
MenReg.fit(Xmens, Ymens)
```

```python
WomenReg = LinearRegression()
WomenReg.fit(Xwomens, Ywomens)
```

```python
TestData = []
for i in range(44):
    TestData.append([0 for i in range(44)])
for i in range(len(TestData)):
    TestData[i][i] = 1
    TestData[i].append(2020)
```

```python
Ymens = MenReg.predict(TestData)
Ymens
```

```python
Ywomens = WomenReg.predict(TestData)
Ywomens
```

```python
country = dfmensfinal.columns.tolist()
country.pop(len(dfmensfinal.columns)-1)
```

```python
dic = {}
for i in range(len(Ymens)):
    l1 = []
    l1.append(Ymens[i])
    l1.append(Ywomens[i])
    dic[country[i]] = l1 

print(dic)
```

```python
import matplotlib.pyplot as plt
```

```python
keys = dic.keys()

keysList = []
for key in dic.keys():
    keysList.append(key)
keysList

values = Ymens
plt.figure(figsize=(13, 13))
plt.xticks(rotation=90)
plt.title("Mens",fontsize=40,)
plt.plot(keysList, values)
```

```python
values = Ywomens
plt.figure(figsize=(13, 13))
plt.xticks(rotation=90)
plt.title("Womens",fontsize=40,)
plt.plot(keysList, values)
```
